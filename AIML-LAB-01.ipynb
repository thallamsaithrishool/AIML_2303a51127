```python
import numpy as np
list1 = [1,2,3]
array1 = np.array(list1)
array1
```

```python
list2 = [3,2,1]
array2=np.array(list2)
array2
```

```python
print("array2 multiplied by array:",array2*array1)
print("array2 divied by :",array2/array1)
print("array2 raised to the power of array1:",array2**array1)
print(f"{array2} raised to the power of {array1}:,{array2}")
```

```python

#sine function
print("Sine :",np.sin(array1))
#logaritham
print("natural logaritham:",np.log(array1))
print("Base-10 logaritham:",np.log10(array1))
print("Base-2 logaritham :",np.log2(array1))
#exponential
print("Exponential:",np.exp(array1))
```

```python
people = ['sai','pavan','ganesh','aneesh','sumanth','nagi']
weight=[55,60,66,46,60,80]
height=[160,150,143,155,130,141]
age=[19,18,20,17,17,18]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("relationship between weight and height of people:")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()
plt.scatter(age,height)
plt.title("relationship between age and height of people:")
plt.ylabel("height in meters")
plt.xlabel("age of patients in years")
```

```python
plt.figure(figsize=(18,5))
plt.title("people weight in kgs",fontsize=16,fontstyle='italic')
#main plot funtion 'bar'
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
#plt.xticks(fontsize=14,rotation=30)
#plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()

```

```python
plt.figure(figsize=(7,16))
plt.title("People weight in kgs",fontsize=18,fontstyle='italic')
#Main plot function 'bar'
plt.bar(x=people,height=weight,width=0.8,color="green",edgecolor='k',alpha=0.8)
plt.xlabel("People",fontsize=16)
#plt.xticks(fontsize=14,rotation=30)
#plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```
